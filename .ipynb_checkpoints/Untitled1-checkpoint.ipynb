```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/Users/deepankjuneja/Desktop/datascienceproject/salary_data_cleaned.csv")
df.head(1)
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,4))
axes[0].hist(df["avg_salary"],bins=30,alpha=0.7,label="Feature-1")
axes[0].hist(df["age"],bins=30,alpha=0.7,label="Feature-2")
axes[0].legend()
axes[0].set_title("Feature distribution")

for label, group in df.groupby("label"):
    axes[1].scatter(group["Feature_1"], group["Feature_2"], label=label, alpha=0.6)
    axes[1].legend()
    axes[1].set_title("Feature relationship by class")


# df["label"].value_counts().plot(kind="bar", ax=axes[2])
# axes[2].set_title("Class balance")

plt.tight_layout()
plt.show()
```

```python
for label, group in df.groupby("label"):
    axes[1].scatter(group["feature_1"], group["feature_2"], label=label, alpha=0.6)
axes[1].legend()
axes[1].set_title("Feature relationship by class")


df["label"].value_counts().plot(kind="bar", ax=axes[2])
axes[2].set_title("Class balance")

plt.tight_layout()
plt.show()
```
